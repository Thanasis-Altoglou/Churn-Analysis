## Creating Cohorts based on Tenure
### Let's begin by treating each unique tenure length, 1 month, 2 month, 3 month...N months as its own cohort.

**Treating each unique tenure group as a cohort, we calculate the Churn rate (percentage that had Yes Churn) per cohort. For example, the cohort that has had a tenure of 1 month should have a Churn rate of 61.99%. We should have cohorts 1-72 months with a general trend of the longer the tenure of the cohort, the less of a churn rate. This makes sense as customers are less likely to stop service the longer they've had it.**

```python
#groupby churn and tenure,then we count the number of instances of people that did and did not churn
df.groupby(['Churn','tenure']).count().transpose()
```

```python
#We grab the churn column
yes_churn = df.groupby(['Churn','tenure']).count().transpose()['Yes']
no_churn = df.groupby(['Churn','tenure']).count().transpose()['No']
```

```python
#calculate the percentage of people who did churn
churn_rate = 100*yes_churn/(yes_churn + no_churn)
```

```python
churn_rate.transpose()['customerID'].plot()
plt.title('Churn Percentage by tenure')
plt.ylabel('Churn rate')
```

According to the line graph, customers who have a tenure of 1 to 3 months are more likely to churn. As the tenure increases, the churn percentage shrinks. The lowest percentage can be observed at 72 months tenure as customers are less likely to churn the longer they stay in a service.                                                                                                                                                           

### Broader Cohort Groups
**Based on the tenure column values, create a new column called Tenure Cohort that creates 4 separate categories:**
   * '0-12 Months'
   * '12-24 Months'
   * '24-48 Months'
   * 'Over 48 Months'

```python
def cohort(tenure):
    if tenure < 13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49:
        return '24-48 Months'
    else:
        return 'Over 48 months'
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df[['tenure','Tenure Cohort']]
```

```python
#scatterplot of Total Charges versus Monthly Charges,colored by Tenure Cohort
plt.figure(figsize=(10,4),dpi=200)
sns.scatterplot(data=df,x='MonthlyCharges',y='TotalCharges',hue='Tenure Cohort')
```

```python
#count plot showing the churn count per cohort.
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='Tenure Cohort',hue='Churn')
```

It's clear again that customers who stay more than 24 months in a service did not churn. However, the amount of people who did churn and those who did not is similar throughout a year.

```python
#grid of Count Plots showing counts per Tenure Cohort, separated out by contract type and colored by the Churn hue.
sns.catplot(data=df,x='Tenure Cohort',hue='Churn',kind='count',
           col='Contract')
```


# Part 5: Predictive Modeling

**Let's explore 4 different tree based methods: A Single Decision Tree, Random Forest, AdaBoost, Gradient Boosting.**


## Single Decision Tree

```python
df.columns
```

```python
#customerID column is unique for every single row, so this feature is not necessary and should be dropped.
#we separate out the data into X feature and Y label

X = df.drop(['Churn','customerID'],axis=1)
X = pd.get_dummies(X,drop_first=True)
```

```python
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier #import of decision tree model
from sklearn.metrics import accuracy_score,classification_report,plot_confusion_matrix #import of the evaluation metrics
```

```python
dt = DecisionTreeClassifier(max_depth=6)
```

```python
#fit the model on the training data in order to fairly evaluate our model's performance on future data
dt.fit(X_train,y_train) 
```

```python
#evaluate performance on test set (predictions)
preds = dt.predict(X_test)
```

```python
#print the classification report
print(classification_report(y_test,preds))
```

```python
accuracy_score(y_test,preds) #model's performance
```

```python
plot_confusion_matrix(dt,X_test,y_test)
```

```python
#calculate the feature importance
dt.feature_importances_
```

```python
imp_feats = pd.DataFrame(data = dt.feature_importances_,index=X.columns,
                        columns=['Feat Imp'])
```

```python
#imp_feats.sort_values('Feat Imp')
```

```python
imp_feats = imp_feats[imp_feats['Feat Imp']>0] 
```

```python
imp_feats = imp_feats.sort_values('Feat Imp')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.barplot(data=imp_feats,x=imp_feats.index,y='Feat Imp')
plt.xticks(rotation=90)
plt.title('Feature Importance for Decision Tree')
```

The most importance feature is tenure, so as the InternetService_Fiber optic. Therefore that is a really important feature that wan't too clear from our previous analysis.

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(max_depth=6)
```

```python
#fit the model on the training data in order to fairly evaluate our model's performance on future data
rf.fit(X_train,y_train)
```

```python
preds = rf.predict(X_test)
```

```python
print(classification_report(y_test,preds))
```

```python
plot_confusion_matrix(rf,X_test,y_test)
```
